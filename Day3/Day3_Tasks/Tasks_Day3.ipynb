### 6. Write a custom function that removes rows where more than 30% of data is missing.

```python
# Function to remove the row, where more than 30 percent of data in column is missing
def remove_row(p):
    for item in range(0,len(p)+1):
        per=(p.loc[item].isnull().sum())/len(p.loc[item])    
        if per >= 0.3:
            p=p.drop(p.index[item])

        return p
```
