```python
from sklearn.preprocessing import OneHotEncoder

OH_encoding = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
OH_cols_train = pd.DataFrame(OH_encoding.fit_transform(x_train[list_cat]))
OH_cols_valid = pd.DataFrame(OH_encoding.transform(x_valid[list_cat]))
```

```python
OH_cols_train.index=x_train.index
OH_cols_valid.index=x_valid.index

num_x_train=x_train.drop(list_cat,axis=1)
num_x_valid=x_valid.drop(list_cat,axis=1)

oh_x_train=pd.concat([num_x_train,OH_cols_train],axis=1)
oh_x_valid=pd.concat([num_x_valid,OH_cols_valid],axis=1)

oh_x_train.columns=oh_x_train.columns.astype(str)
oh_x_valid.columns=oh_x_valid.columns.astype(str)
```

### Data Imputation

```python
# 1: Groupby Method: Fill the null values with similar data/value

car_data['Engine Fuel Type']=car_data.groupby('Model')['Engine Fuel Type'].transform(lambda x:x.fillna(x.mode().iloc[0] if not x.mode().empty else None))
```

```python
# Fill missing values in numeric columns using group-wise mean

car_data['Number of Doors'] = car_data.groupby('Model')['Number of Doors'].transform(lambda x: x.fillna(x.mean()))
car_data['Engine Cylinders'] = car_data.groupby('Model')['Engine Cylinders'].transform(lambda x: x.fillna(x.mean()))
car_data['Engine HP'] = car_data.groupby('Model')['Engine HP'].transform(lambda x: x.fillna(x.mean()))
```

```python
car_data.isnull().sum()
```

```python
print(car_data['Engine Cylinders'].unique())
```

```python
null_data=car_data[car_data['Engine Cylinders'].isnull()]
null_data
```

```python
# for electric car, number of cylinders should be 0
car_data.loc[(car_data['Engine Cylinders'].isna() & (car_data['Engine Fuel Type']=='electric'), 'Engine Cylinders')]=0
```

```python
car_data.loc[(car_data['Engine Cylinders'].isna() & (car_data['Make']=='Mazda'), 'Engine Cylinders')]=4
```
