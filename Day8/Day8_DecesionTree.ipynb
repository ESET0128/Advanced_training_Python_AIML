## Decision Tree : Iris Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
data=load_iris()
data
```

```python
data['DESCR']
```

```python
x=pd.DataFrame(data['data'])
y=pd.DataFrame(data['target'])
```

```python
sns.boxplot(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classifier=DecisionTreeClassifier()
tree_classifier.fit(x_train,y_train)
```

```python
y_pred=tree_classifier.predict(x_test)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_classifier,filled=True)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print("Confusion Matrix:\n",confusion_matrix(y_test,y_pred))
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
print("Train accuracy_score:",accuracy_score(y_train,tree_classifier.predict(x_train)))
print("Test accuracy_score:",accuracy_score(y_test,y_pred))
```

## Decision Tree : Diabetes Dataset

```python
from sklearn.datasets import load_diabetes
data_d=load_diabetes()
data_d
```

```python
data_d['DESCR']
```

```python
xd=pd.DataFrame(data_d['data'])
yd=pd.DataFrame(data_d['target'])
```

```python
xd.corr()
```

```python
sns.heatmap(xd.corr(),annot=True)
```

```python
sns.boxplot(xd)
```

```python
xd_train,xd_test,yd_train,yd_test=train_test_split(xd,yd,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
reg.fit(xd_train,yd_train)
```

```python
yd_pred=reg.predict(xd_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
r_score=r2_score(yd_test,yd_pred)
print("R2_Score:",r_score)
mean_abs_err=mean_absolute_error(yd_test,yd_pred)
print("Mean Absolute Error",mean_abs_err)
mean_sq_err=mean_squared_error(yd_test,yd_pred)
print("Mean Squared Error",mean_sq_err)

```
