### Taking only that data rows where (20 <= purchase_amount <= 2000)

### Cheking preferred_product_category and setting as per requirement

```python
data['preferred_product_category'].unique()
```

```python
data=data[data['preferred_product_category'].isin(['Electronics', 'Clothing', 'Home & Garden', 'Toys','Health'])]
```

### Selecting customer_loyalty_score between 0 and 100

```python
data=data[(data['customer_loyalty_score'] >=0) & (data['customer_loyalty_score'] <=100)]
```

### Setting up signup_date between '2021-01-01' and '2022-12-31'

```python
data = data[data['signup_date'].between('2021-01-01', '2022-12-31')]
```

```python
data['signup_date'] = pd.to_datetime(data['signup_date'])
data['last_purchase_date'] = pd.to_datetime(data['last_purchase_date'])
data = data[data['last_purchase_date'] > data['signup_date']]
```

```python
monthly2 = data.set_index('last_purchase_date').index.to_period('M')
s2 = pd.Series(1, index=monthly2).groupby(level=0).sum().sort_index()
s2 = s2.rename_axis('month').astype(float)

plt.figure()
s2.index = s2.index.astype(str)
plt.plot(s2.index, s2.values, marker='o')
plt.title("Monthly Last-Purchase Events")
plt.xlabel("Month")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')

```
